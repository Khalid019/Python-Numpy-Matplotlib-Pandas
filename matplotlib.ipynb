## Matplotlib

#### Matplotlib is ploting libraray for the Python programming language. It is a comprehensive library for creating static, animated, and interactive visualizations in Pyhton.

### Import

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Created two Data sets 

```python
x_data = [1, 2, 3, 4, 6, 8, 9, 10, 11, 12]
y_data = [4000, 6000, 6500, 7000, 7100, 7700, 7800, 8000, 8100, 8200]
```

### Plot using plt 

```python
plt.plot(x_data, y_data)
```

### Adding axis label and tilte of the plot

```python
plt.plot(x_data, y_data)
plt.xlabel("Day")
plt.ylabel("Income")
plt.title("Income per day")
```

### Creating another data set of y

```python

y_data_2 = [6000, 6400, 6800, 7500, 7800, 7900, 8000, 8500, 8700, 8900]
```

### Plot

```python
# first data set
plt.plot(x_data, y_data)

# second data set
plt.plot(x_data, y_data_2)

plt.xlabel("Day")
plt.ylabel("Income")
plt.title("Income per day")
```

### Add legend on the plot

```python
# first data set
plt.plot(x_data, y_data, label="first dataset")

# second data set
plt.plot(x_data, y_data_2, label="second dataset")

plt.xlabel("Day")
plt.ylabel("Income")
plt.title("Income per day")
plt.legend() # add legend
```

### Adding color on the plot

```python
# first data set
plt.plot(x_data, y_data, 'r--' ,label="first dataset")

# second data set
plt.plot(x_data, y_data_2,'g', label="second dataset")

plt.xlabel("Day")
plt.ylabel("Income")
plt.title("Income per day")
plt.legend() # add legend
```

### Use marker on the plot

```python
# first data set
plt.plot(x_data, y_data, 'r--', marker= '*',label="first dataset")

# second data set
plt.plot(x_data, y_data_2,'g', marker= 'o',label="second dataset")

plt.xlabel("Day")
plt.ylabel("Income")
plt.title("Income per day")
plt.legend() # add legend
```

### Adding linewidth on the plot

```python
# first data set
plt.plot(x_data, y_data, 'r--', marker= '*', linewidth=2,label="first dataset")

# second data set
plt.plot(x_data, y_data_2,'g', marker= 'o', linewidth=2, label="second dataset")

plt.xlabel("Day")
plt.ylabel("Income")
plt.title("Income per day")
plt.legend() # add legend
```

### Adding grid on the figure

```python
# first data set
plt.plot(x_data, y_data, 'r--', marker= '*',label="first dataset")

# second data set
plt.plot(x_data, y_data_2,'g', marker= 'o',label="second dataset")

plt.xlabel("Day")
plt.ylabel("Income")
plt.title("Income per day")
plt.legend() # add legend
plt.grid(True) # add grid here
```

### Saving Figure

```python
# first data set
plt.plot(x_data, y_data, 'r--', marker= '*',label="first dataset")

# second data set
plt.plot(x_data, y_data_2,'g', marker= 'o',label="second dataset")

plt.xlabel("Day")
plt.ylabel("Income")
plt.title("Income per day")
plt.legend() # add legend
plt.savefig('matplotlib.png')
```
